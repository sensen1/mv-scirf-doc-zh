示例 2：测量 190 GHz 有源双端口器件。

以下示例展示了如何使用矢量拟合法，使用从 140 GHz 到 220 GHz 测量的有源电路的 2 端口散射参数矩阵。更多说明和背景信息请参考 [矢量拟合教程](../../tutorials/VectorFitting.ipynb)。

```python
import numpy as np

import skrf
from skrf.vectorFitting import VectorFitting
```

这个例子更难拟合，因为响应中包含一些“峰值”和测量噪声。在这种情况下，找到合适的初始极点数可能需要进行几次迭代。对于这种情况，一个好的替代方案是自动矢量拟合功能 `auto_fit()`，它可以自动调整极点数，直到获得良好的拟合结果。

从 Touchstone 文件中加载网络数据，创建矢量拟合实例，并启动自动拟合：

```python
nw = skrf.network.Network('./190ghz_tx_measured.S2P')
vf = VectorFitting(nw)
vf.auto_fit()
```

`plot_convergence()` 函数可以帮助检查收敛情况，并查看是否存在问题。需要检查的重要参数包括达到收敛所需的迭代次数以及最终模型阶数。

```python
vf.plot_convergence()
```

```python
print(f'model order = {vf.get_model_order(vf.poles)}')
print(f'n_poles_real = {np.sum(vf.poles.imag == 0.0)}')
print(f'n_poles_complex = {np.sum(vf.poles.imag > 0.0)}')
```

通过将模型响应与原始采样数据进行比较来验证结果，表明拟合成功，这一点也体现在较小的均方根误差（小于 0.05）上：

```python
vf.get_rms_error()
```

```python
# plot frequency responses
vf.plot_s_db()
```

最好也检查模型在原始频率范围之外的响应情况。

```python
freqs = np.linspace(0, 500e9, 501) # plot model response from dc to 500 GHz
vf.plot_s_db(freqs=freqs)
```
