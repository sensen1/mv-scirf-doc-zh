# TRL

这个简单的示例演示了如何使用 scikit-rf 进行基本的 TRL 校准。

## 基本用法### 导入首先，我们导入 `skrf` 库、`TRL` 类，并设置 matplotlib 绘图环境。

```python
import skrf as rf
from skrf.calibration import TRL

%matplotlib inline
from skrf.plotting import stylely

stylely()
```

### 加载数据

从 Touchstone 文件中加载校准标准的测量数据和开关参数。（如果您不清楚什么是开关参数，请参阅下面的[此部分](#Switch-Terms)。）

```python
T = rf.Network('trl_data/thru.s2p')
R = rf.Network('trl_data/reflect.s2p')
L = rf.Network('trl_data/line.s2p')

switch_terms = (rf.Network('trl_data/forward switch term.s1p'),
                rf.Network('trl_data/reverse switch term.s1p'))

measured = [T,R,L]
```

### 创建一个 TRL 校准对象

```python
trl = TRL(measured = measured,
          switch_terms = switch_terms)
```

使用它校正待测器件。

```python
dut_raw = rf.Network('trl_data/mismatched line.s2p')
dut_corrected = trl.apply_cal(dut_raw)
dut_corrected.plot_s_db()
```

## 开关术语

### 它们是什么？如 Roger Marks 在 [1] 中所述，开关项考虑了 8 项（也称为误差盒）模型过于简化的事实。这两个误差网络会根据激励哪个端口而略有变化。这是由于 VNA 内部的开关造成的。VNA 内部开关处的反射系数变化被称为 *开关项*，它可以直接测量并消除。* [1] R.B. Marks，《包括开关项的基本 VNA 误差模型的公式》，载于 ARFTG 会议文集——秋季，第 50 届，1997 年，第 32 卷，第 115-126 页。### 我如何获取它们？要手动测量开关项，您可以定义自定义迹线，使其为接收器的比率，并将数据保存到 Touchstone 文件中。假设端口标记为 1 和 2，入射波为 a，反射波为 b。那么：* **正向开关项** == `a2/b2`，源端口为 1* **反向开关项** == `a1/b1`，源端口为 2### 我可以使用 skrf 获得它们吗？是的。`skrf.vi.vna` 模块中有一个 PNA 驱动程序。它包含用于提取开关项的代码。### 我可以在 PNA 上获取它们吗？是的。请参见下面的图像，了解在 PNA 上设置此功能的示例。<img src="switch_term_dialog.png" width="800"/>
