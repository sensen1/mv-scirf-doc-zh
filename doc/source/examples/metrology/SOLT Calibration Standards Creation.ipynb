# 创建 SOLT 校准标准

## 简介在 scikit-rf 中，校准标准被视为一个普通的单端口或双端口 `skrf.Network`，其定义方式是基于其完整的 S 散射参数。它可以表示反射、传输、负载，甚至任意阻抗的标准。由于在电路建模和拟合过程中不会引入额外的误差，因此这种方法可以实现最高的校准精度，并且在 VNA 厂商的术语中被称为“基于数据的标准”。但是，VNA 校准标准通常使用具有拟合系数的电路模型来定义。由于基于数据的标准通常作为优质产品提供，因此许多基于模型的校准套件仍在继续使用和制造（甚至在 50 GHz 时 [[1](SOLT%20Calibration%20Standards%20Creation.html#ref1)]）。这需要在将其用于 scikit-rf 的校准例程之前创建它们的网络模型。本示例解释了如何从 HP-Agilent-Keysight 格式和 Rohde & Schwarz / Anritsu 格式给出的系数中创建网络。两者本质上是相同的电路模型，但后者格式对偏移传输线使用不同的测量单位。Keysight 85032F（Type-N 插件）、Keysight 85033E（3.5 mm 插件）、通用平面标准（SMA 插件）和 Maury Microwave 8050CK10（3.5 mm 插件）的计算响应也以示例的形式记录在本文档中。.. 注意：本指南仅涵盖同轴标准。对于波导，计算方法不同。特别是，对于同轴线，不能应用 $\sqrt{\text{GHz}}$ 的缩放，因为损耗也是其物理尺寸的函数，并且公式更加复杂。您是否有波导方面的经验？如果有，您可以帮助通过[贡献](../../contributing/index.rst#examples-and-tutorials) 到文档。## scikit-rf 建模的替代方案在我们开始之前，值得指出一些替代方案。在 scikit-rf 中，您可以使用任何现有的标准定义，只需提供其 S 散射参数。如果您已经将标准定义为其他工具（例如您喜欢的电路仿真器或实际测量结果）中的网络，则可以简单地将 S 散射参数导出到 Touchstone 文件中，以便在 scikit-rf 中使用。同样，如果您已经在使用基于数据的校准标准，则应该可以直接使用其数据。S 散射参数可以存储在特定于设备的的文件格式中，请咨询您的供应商，了解是否可以将其导出为 Touchstone 文件。作为一种特殊情况，如果使用“平面”（零长度）标准，这意味着标准直接位于连接器的参考平面上，而没有延伸的主体，那么有时可以假定校准标准对于非关键测量是理想的。在 scikit-rf 中，可以通过定义理想传输线并调用 `short()`、`open()`、`match()` 和 `thru()` 方法（在 [准备](#preparation) 部分中说明）来方便地创建理想响应。.. 重要：理想假设仅对“平面”标准在一定程度上有效。大多数实验室级标准都是“偏移”标准，请始终使用正确的定义来使用这些标准。它们具有由其延伸主体产生的额外电延迟，如果采用理想假设，相位误差将不可接受地大。在“直通”适配器后面使用“平面”标准会产生相同的相位偏差。请参见下文以了解“平面”标准和“偏移”标准之间的区别。## HP-Agilent-Keysight ### 电路模型在开始创建它们的网络定义之前，我们首先需要了解底层的电路模型以及这些系数的含义。如图所示，这是校准标准的 HP-Agilent-Keysight 模型。<figure><img src="solt_calibration_standards_creation/calkit-schematic.svg" width="40%"><figcaption>*校准标准的电路模型*</figcaption></figure>#### 终端阻抗在 *偏移* 传输线的末端连接了一个并联阻抗，用于模拟开路或短路标准中的分布式电容或电感。它表示为一个三阶多项式，具有四个系数，$y(f) = a_0 + a_1 f + a_2 f^2 + a_3 f^3$，其中 $f$ 是频率，$a_i$ 是系数。对于开路标准，它们是 $\text{C}_0$、$\text{C}_1$、$\text{C}_2$、$\text{C}_3$，第一个常数项以飞法 (femtofarad) 为单位。对于短路标准，它们是 $\text{L}_0$、$\text{L}_1$、$\text{L}_2$、$\text{L}_3$，第一个常数项以皮亨 (picohenry) 为单位。.. important::  终端阻抗的反射系数 ($S_{11}$) 相对于系统阻抗 ($Z_\mathrm{ref}$) 进行计算，而不是相对于偏移传输线的阻抗 ($Z_\text{off}$ 或 $Z_\text{c}$) 进行计算。  [[15](SOLT%20Calibration%20Standards%20Creation.html#ref15)]#### 偏移线在校准标准的前面，存在一个 *偏移* 损耗传输线。为了理解其含义，我们需要区分两种类型的校准标准：*紧贴*（零长度）和 *偏移* 标准。在 *紧贴* 标准中，物理标准直接位于端口的参考平面上。这种设计广泛用于未表征的通用 SMA 校准“标准”（如果可以这样称呼的话），应用于业余无线电中。*紧贴* 标准并不总是质量差，也可以在一些实验室级别的标准中找到，例如 Keysight 85031B，用于 APC-7 连接器。

<figure><img src="solt_calibration_standards_creation/flush-calkit.jpg" width="45%"><img src="solt_calibration_standards_creation/offset-calkit.jpg" width="45%"><figcaption>*齐平校准标准与连接器紧密贴合，没有额外的结构。 偏移式校准标准具有延伸的本体，引入电延迟。*</figcaption></figure>然而，大多数实验室级标准都设计为*偏移式*标准，连接器参考平面和标准物理位置之间具有较短的长度。 这样做是为了提高标准电气特性的可预测性 [[2](SOLT%20Calibration%20Standards%20Creation.html#ref2)]，或者仅仅是因为连接器的机械设计存在限制（例如，所有 N 型连接器都具有 5.258-5.360 mm 的参考平面偏移量 [[3](SOLT%20Calibration%20Standards%20Creation.html#ref3)]）。 这两个因素都会引入必须考虑的电长度，否则会产生很大的相位偏差。这种*偏移*长度被建模为一个传输线，使用三个参数：1. **无损偏移阻抗** ($Z_\text{off}$)：一个实数特性阻抗，忽略了线路损耗（R 和 G）。   这通常与 VNA 的参考阻抗匹配，但并非总是如此。 有时，会使用与参考阻抗略有不同的值来模拟物理缺陷，例如 50.209 Ω 或 49.992 Ω。   此外，波导标准使用特殊的归一化值 `1`。2. **偏移延迟** ($t_\text{delay}$)：单向电延迟，以皮秒 (ps) 为单位。3. **偏移损耗** ($A_\text{loss}$)。在 1 GHz 时的标称衰减，单位为 GΩ/s（十亿欧姆/秒）。将此值按   $\sqrt{f / (\text{1 GHz})}$ 缩放，用于其他频率。   GΩ/s 单位虽然不常见，但有其原因。为了表达传输线段的长度，可以使用距离（以米为单位）或时间延迟（以秒为单位）。将线路定义为时间相关的，可以计算   $R = A_\text{loss} t_\text{delay}$，单位为欧姆，而无需显式定义介电常数 $\epsilon$。.. 重要提示::    这些参数在数据表中以缩放单位给出，但在所有计算中应使用未缩放的 SI 单位（无前缀）：欧姆 (Ω)、秒 (s)、欧姆/秒 (Ω/s) 和赫兹 (Hz)。##### RLCG 参数<figure><img src="solt_calibration_standards_creation/rlcg.svg" width="40%"><figcaption>*RLCG 传输线模型*</figcaption></figure>由于这些参数不是标准参数，因此我们需要将它们转换为熟悉的 RLCG 线路参数。 在经过 [[4](SOLT%20Calibration%20Standards%20Creation.html#ref4)] 方程 2B.4 的代数运算后，相应的与频率相关的 RLCG 参数计算如下。$$\begin{aligned}R(f) \cdot l &= \left(A_\text{loss} t_\text{delay}\right) \cdot \sqrt{\frac{f}{10^9}} \\L(f) &= L_0 + L_\text{cond} = t_\text{delay}  Z_\text{off} + \dfrac{R}{2\pi f} \\C &= \dfrac{t_\text{delay}}{Z_\text{off}} \\G &= 0 \\l &= 1\text{ (归一化)}\end{aligned}$$$L_0$ 项是理想线路电感，$L_\text{cond} = R / (2 \pi f)$ 是由于趋肤效应引起的具有频率依赖性的不完全导体电感。 在 [[5](SOLT%20Calibration%20Standards%20Creation.html#ref5)] 中，推导出了几乎相同的方程，只是省略了 $L_\text{cond}$ 项。 为了与原始 Keysight 发布的文档 [[4](SOLT%20Calibration%20Standards%20Creation.html#ref4)] 保持一致，必须在此处包含该项（否则，在 9 GHz 时，85032F 的相位误差将大于 0.1 度）。使用这些参数，可以计算出有损线路的*传播常数* ($\gamma$) 和复数*特性阻抗* ($Z_c$)。 请注意，在与频率相关的 `RLCG(f)` 模型中，每个频率都有其自身的 RLCG 参数和线路常数。.. 重要提示:: 我们的公式与官方 Keysight 计算和其他工作不完全相同。 后者在计算线路常数时使用了低损耗近似，而我们的公式是从 RLCG 参数开始推导出精确的线路常数。 两种公式的差异可以忽略不计（Keysight 85056A 在 50 GHz 时的幅度和相位角在小数点后 4 位上一致），这种差异仅是学术上的。.. 注意:: 我们展示了 RLCG 方法，因为它具有广泛的理解性和易用性。 如果绝对需要再现到最后一位数字的精确数值，请参见*附录*以获取 Keysight 官方低损耗近似的推导和代码。#### 参考阻抗所有校准标准都具有隐含的*参考阻抗* $Z_\text{ref}$。 一些 Keysight 数据表将其显示为“System $Z_0$”[[6](SOLT%20Calibration%20Standards%20Creation.html#ref6)]，但并非总是明确给出。 可能需要一些上下文来推断它：如果一个标准命名为*Keysight 85032F Type N (50) 校准套件*，则 $Z_\text{ref}$ 为 50 Ω。不要将其与偏移传输线的特性阻抗 $Z_0$（无损）或 $Z_c$（有损）相混淆，后者可以具有 49.992 Ω 的值。 与*线路阻抗*不同，*参考阻抗*的目的是将散射参数重新归一化为标准形式，因为所有 S 参数都是相对的。 这始终设置为 VNA 系统的参考阻抗：50 Ω 或 75 Ω。##### 连接器不连续性由于参考阻抗似乎是一个任意选择，因此有人可能会得出结论，可以使用任何校准标准来校准任何 VNA 端口（只要它们在机械上兼容），而不管其偏移线路阻抗如何。 例如，可以尝试通过将其直接连接到 75 Ω 校准标准来校准 50 Ω VNA 端口，以便所有校准的测量值都表现得好像是使用 75 Ω 系统进行测量的。 同样，如果重新归一化 75 Ω 校准标准的参考阻抗，似乎可以预测其在 50 Ω VNA 上的理想 S 参数，从而将其用作 50 Ω 标准。但是，这个假设是不正确的。 重新归一化无法预测由于传输线物理尺寸突然变化而引起的物理测试端口的寄生效应，即电磁场的间断性。 [[7](SOLT%20Calibration%20Standards%20Creation.html#ref7)][[9](SOLT%20Calibration%20Standards%20Creation.html#ref9)] 大致来说，这是一个容性不连续性。 [[18](SOLT%20Calibration%20Standards%20Creation.html#ref18)] 即使两端具有相同的特性阻抗，只要同轴线的物理尺寸发生阶跃变化（例如，从 3.5 mm 到 2.92 mm 端口），也会出现这种效应。 [[8](SOLT%20Calibration%20Standards%20Creation.html#ref8)]从理论上讲，不连续性处的高阶模式的激发会使校准不可靠，因为它可能对扰动敏感且不可重复。 事实上，即使在两个“完美”的 1.85 mm 连接器之间，也观察到了病态情况——匹配的连接器对会产生共振，因为其机械公差*过于精确*，近乎零的间隙形成了一个谐振器。 [[25](SOLT%20Calibration%20Standards%20Creation.html#ref25)][[26](SOLT%20Calibration%20Standards%20Creation.html#ref26)]然而，在大多数不受此类病态现象影响的实际测量中，已经证明，如果 DUT 和校准标准具有相同的连接器设计，则它们都会产生类似的不连续性，并且通常可以将这种不连续性校准掉，从而只留下一个对于所有但最苛刻的计量应用而言都可以忽略不计的微小误差。 [[24](SOLT%20Calibration%20Standards%20Creation.html#ref24)] 当使用 50 Ω 端口校准 75 Ω 设备并使用 75 Ω 标准时，这就是这种情况。 [[9](SOLT%20Calibration%20Standards%20Creation.html#ref9)] 但是，如果 DUT 和校准标准具有不同的连接器设计（例如，交叉连接 3.5 mm、2.92 mm 和 SMA），则这种不连续性会产生轻微的测量误差。 [[8](SOLT%20Calibration%20Standards%20Creation.html#ref8)]因此，为了获得令人满意的结果，校准标准和 DUT 必须使用相同的连接器。 为了获得最佳结果，测试端口（或适配器）必须具有高质量，并且与校准标准相匹配。#### 拟合非唯一性和物理可解释性小电容或电感的相位变化大致是线性的，通常可以将其建模为终止电抗，也可以建模为偏移传输线中的电延迟。 [[4](SOLT%20Calibration%20Standards%20Creation.html#ref4)][[10](SOLT%20Calibration%20Standards%20Creation.html#ref10)]例如，Keysight 85033D 的数据表显示了短路标准的两种模型，一种是纯时间延迟，另一种是具有电感的延迟。 [[6](SOLT%20Calibration%20Standards%20Creation.html#ref6)]由于标准定义的非唯一性，在解释标准定义中的各个术语时，必须小心：整体响应是有意义的，各个术语可能是有意义的，也可能没有意义。 大多数实验室级校准标准都试图确保标准定义在物理上是有意义的 [[11](SOLT%20Calibration%20Standards%20Creation.html#ref11)]，但这并非总是得到保证。在使用曲线拟合定义自定义校准标准时，将由于偏移传输线和终止电抗引起的相位变化分开是很困难的。 通过先验知识，可以通过机械测量来固定偏移长度 [[12](SOLT%20Calibration%20Standards%20Creation.html#ref12)]，或者通过电磁理论或模拟来固定终止电容。 [[10](SOLT%20Calibration%20Standards%20Creation.html#ref10)] 这两者都需要对标准的設計进行精确的了解和建模。 例如，历史研究侧重于空心空气填充的开路标准（使用截止频率以下的圆形波导理论）[[10](SOLT%20Calibration%20Standards%20Creation.html#ref10)]，但今天不再适用，因为许多开路标准使用介电支撑。 当缺乏信息时，仅延迟模型可能是唯一的实际选择。.. 提示::    实际上，仅延迟的短路标准是较旧的 VNA 固件所支持的唯一模型，例如 HP 8510 [[13](SOLT%20Calibration%20Standards%20Creation.html#ref13)]。 通过 scikit-rf 进行外部校准，您可以绕过此限制。#### 忽略的项开路或短路都可以建模为仅具有电延迟，而没有电抗，反之亦然。 仅延迟模型最常用于短路标准，因为它在历史上一直如此。匹配负载会产生很少的反射，因此通常将其建模为 $Z_0$ 终端。 由于其反射系数约为 0，因此其相位角度没有意义，因此其电延迟也忽略不计。直通标准有时仅建模为具有偏移延迟，如果损耗可以忽略不计。一些校准套件没有提供任何直通延迟数据。 这些套件用于*齐平直通*、*互换等效适配器*或*未知直通*校准，而不使用*直通*的数据。*齐平直通*是传统的 SOLT 校准方法：端口 1 和端口 2 直接连接，不使用适配器。 此*直通*在定义上是理想的，并且具有零长度，不需要建模。 在*互换等效适配器*校准中，使用一对延迟匹配的适配器，从而使*直通*的实际长度变得不可见。 *未知直通*校准算法是相对较新的，不需要对直通进行表征。.. 注意::    有时偏移损耗仍然出现在负载和直通的定义中，但具有零电长度。 这些参数应被忽略。 零偏移延迟禁用了偏移传输线。#### 插头和插座最后，值得阐明 Keysight 的插头和插座符号。 如果使用*插头*标准来校准*插座*，则数据表将其相对于自身表示为 `-m-`，或者相对于测试端口表示为 `(f)`。 如果使用*插座*标准来校准*插头*，则数据表将其相对于自身表示为 `-f-`，或者相对于测试端口表示为 `(m)`。 [[14](SOLT%20Calibration%20Standards%20Creation.html#ref14)]

### 准备<div id="preparation"></div>有了这个电路模型，我们可以开始对校准标准进行建模。首先，我们需要导入一些库定义，并指定计算的频率范围。这里，我们使用了 1 MHz 到 9 GHz，共 1001 个点。您可以根据需要进行调整。我们还定义了一个 `ideal_medium`，其端口阻抗为 $50 \space\Omega$，以便将来进行某些计算。

````python
import numpy as np

import skrf
from skrf.media import DefinedGammaZ0, DistributedCircuit

# reference impedance of the S-parameters (not the offset line)
z0_ref = 50
freq = skrf.Frequency(1, 9000, 1001, "MHz")
ideal_medium = DefinedGammaZ0(frequency=freq, z0=z0_ref)
````

### 理想情况下的响应首先了解一下特殊情况是有帮助的：理想的校准标准可以通过调用 `ideal_medium` 中的 `open()`、`short()`、`match()` 和 `thru()` 方法轻松创建。前三个方法返回一个 1 端口网络。`thru()` 方法返回一个 2 端口网络。

````python
ideal_open  = ideal_medium.open()
ideal_short = ideal_medium.short()
ideal_load  = ideal_medium.match()
ideal_thru  = ideal_medium.thru()
````

### 建模偏移传输线为了正确地建模偏移传输线，应该使用偏移延迟、偏移损耗和偏移 $Z_0$，从而推导出有损线路的 RLCG 参数。在内部，`scikit-rf` 会自动推导出有损线路的 *传播常数* ($\gamma$) 和复数 *特性阻抗* ($Z_c$)。偏移线路参数与 RLCG 线路参数之间的关系已经在上面的讨论中给出了。现在，我们将这些公式转换为代码。.. important:   $\sqrt{\frac{f}{10^9}}$ 这个项用于将线路损耗从标称 1 GHz 值缩放到给定的频率，但这仅适用于同轴线路。

````python
def offset_rlcg(freq, offset_delay, offset_loss, offset_z0):
    r = offset_loss * offset_delay * np.sqrt(freq.f / 1e9)
    l = (offset_delay * offset_z0) + r / (2 * np.pi * freq.f)
    c = offset_delay / offset_z0
    g = 0
    return (r, l, c, g)
````

.. 提示：这里使用了 `numpy` 中的广播功能。量 `r` 和 `l` 都是与频率相关的，因此它们是数组，而不是标量。与其显式地循环遍历每个频率并将它们添加到数组中，不如在这里通过标量与 `numpy.array` 的乘法自动创建数组。我们将继续使用这种技术。定义了函数 `offset_rlcg()` 后，我们可以通过调用该函数来计算开路和短路校准标准的线路常数。

````python
rlcg_open = offset_rlcg(freq, 40.856e-12, 0.93e9, 50)
rlcg_short = offset_rlcg(freq, 45.955e-12, 1.087e9, 49.992)
````

到目前为止，我们已经掌握了关于该移相线的必要信息。任务的另一半很简单：在 scikit-rf 中，使用这些 RLCG 参数创建一个双端口网络，用于表示该传输线。scikit-rf 将帮助我们自动计算其传播常数 $\gamma l$ 和复数特性阻抗 $Z_c$。在 scikit-rf 中执行此任务非常简单：1. 首先，使用四个数组（RLCG）作为输入，创建一个 `DistributedCircuit` 介质。创建的 `DistributedCircuit` 代表一种物理介质。2. 在创建介质时，我们还需要指定 S 参数参考阻抗 $Z_\text{ref}$，如 `z0_port=50`（或 `z0_port=75`）。   * 不要将此参考阻抗 $Z_\text{ref}$ 与移相线阻抗 $Z_\text{off}$ 和 $Z_c$ 混淆。$Z_\text{ref}$（实数）始终是 VNA 的标称系统或端口阻抗，而线阻抗 $Z_\text{off}$（实数）和 $Z_c$（复数）可能因物理缺陷和损耗而有所不同。3. 然后，通过调用介质的 `line()` 方法，可以创建一个实际长度为 1 米的线路。

````python
medium_open = DistributedCircuit(frequency=freq,
    R=rlcg_open[0], L=rlcg_open[1], C=rlcg_open[2], G=rlcg_open[3],
    z0_port=z0_ref
)
line_open = medium_open.line(
    d=1, unit='m'
)

medium_short = DistributedCircuit(frequency=freq,
    R=rlcg_short[0], L=rlcg_short[1], C=rlcg_short[2], G=rlcg_short[3],
    z0_port=z0_ref
)
line_short = medium_short.line(
    d=1, unit='m'
)
````

### 建模分路阻抗接下来，我们需要对开路和短路校准标准的分路阻抗进行建模。对于开路标准，它是一个电容。对于短路标准，它是一个电感。两者都建模为关于频率的三次多项式。在 `numpy` 中，可以通过 `np.poly1d([x3, x2, x1, x0])` 快速定义这样的函数。这是一个高阶函数，它接受一个按降序排列的系数列表，并返回一个可调用的多项式函数。在对多项式进行求值后，我们可以生成频率相关的电容和电感。开路电路建模为一个串联的 `medium.capacitor()`，然后是一个理想的 `medium.short()`。短路电路建模为一个串联的 `medium.inductor()`，然后是一个理想的 `medium.short()`。根据 [[15](SOLT%20Calibration%20Standards%20Creation.html#ref15)]，电容和电感的 S 参数是相对于系统的参考阻抗定义的，而不是 *偏移* 传输线，因此我们使用 `ideal_medium` 以避免混淆。尽管如此，`medium_open` 或 `medium_short` 也是可以接受的：它们也正确地使用 `skrf.Media()` 的端口阻抗作为参考阻抗。

````python
# use ideal_medium, not medium_open and medium_short to avoid
# reference impedance confusions.

capacitor_poly = np.poly1d([
      13.400 * 1e-45,
    -264.990 * 1e-36,
    2536.800 * 1e-27,
      89.939 * 1e-15
])
capacitor_list = capacitor_poly(freq.f)
shunt_open = ideal_medium.capacitor(capacitor_list) ** ideal_medium.short()

inductor_poly = np.poly1d([
      -0.7847 * 1e-42,
      34.8314 * 1e-33,
    -496.4808 * 1e-24,
       3.3998 * 1e-12
])
inductor_list = inductor_poly(freq.f)
shunt_short = ideal_medium.inductor(inductor_list) ** ideal_medium.short()
````

### 关于串联和并联阻抗的说明为了对并联阻抗（例如用于开路校准标准的电容）进行建模，可以使用一个串联的 `medium.capacitor()`，并将其一端连接到 `medium.short()`。```# [端口] ---- [电容] --- [端口]#                             |#                             |#                         [短路]#                             |#                             |#                        GNDshunt_open = ideal_medium.capacitor(capacitor_list) ** ideal_medium.short()```或者，可以使用一个并联的 `medium.shunt_capacitor()`，并将其一端连接到 `medium.open()`。```# [端口] -----v----- [端口] ----- [开路]#             |#             |#     [并联电容]#             |#             |#           [GND]shunt_open = ideal_medium.shunt_capacitor(capacitor_list) ** ideal_medium.open()```两者是等效的。`medium.open()` 端接对于 `shunt_capacitor()` 来说非常重要：它创建了一个带有连接到地线的电容的二端口网络——这个网络类似于示波器端接器，它是一个带有并联 50 Ω 电阻的 2 端口器件——因此，另一个“输出”端口需要是开路的。否则，仅由 `shunt_capacitor()` 端接的线路会产生不正确的 S-parameters。

### 完成

最后，我们将这些模型组件连接在一起，并添加理想负载和直通的定义，这样就完成了我们的建模。

````python
open_std = line_open ** shunt_open
short_std = line_short ** shunt_short
load_std = ideal_medium.match()
thru_std = ideal_medium.thru()
````

### 代码片段为了方便起见，您可以重复使用以下代码片段，根据 Keysight 格式中的系数生成校准标准网络。

````python
import numpy as np

import skrf
from skrf.media import DefinedGammaZ0, DistributedCircuit


def keysight_calkit_offset_line(freq, offset_delay, offset_loss, offset_z0, ref_z0):
    if offset_delay or offset_loss:
        r = offset_loss * offset_delay * np.sqrt(freq.f / 1e9)
        l = (offset_delay * offset_z0) + r / (2 * np.pi * freq.f)
        c = offset_delay / offset_z0
        g = 0

        medium = DistributedCircuit(frequency=freq,
            R=r, L=l, C=c, G=g,
            z0_port=ref_z0
        )
        offset_line = medium.line(d=1, unit='m')
        return medium, offset_line
    else:
        medium = DefinedGammaZ0(frequency=freq, z0=ref_z0)
        line = medium.line(d=0)
        return medium, line


def keysight_calkit_open(freq, offset_delay, offset_loss, c0, c1, c2, c3, offset_z0, ref_z0):
    # Capacitance is defined with respect to the system reference impedance ref_z0, not the
    # lossy line impedance. In scikit-rf, the return values of `shunt_capacitor()` and
    # `medium.open()` methods are (correctly) referenced to z0_port, which has been set to
    # ref_z0.
    medium, line = keysight_calkit_offset_line(freq, offset_delay, offset_loss, offset_z0, ref_z0)
    if c0 or c1 or c2 or c3:
        poly = np.poly1d([c3, c2, c1, c0])
        capacitance = medium.shunt_capacitor(poly(freq.f)) ** medium.open()
    else:
        capacitance = medium.open()
    return line ** capacitance


def keysight_calkit_short(freq, offset_delay, offset_loss, l0, l1, l2, l3, offset_z0, ref_z0):
    # Inductance is defined with respect to the system reference impedance ref_z0, not the
    # lossy line impedance. In scikit-rf, the return values of `shunt_inductance()` and
    # `medium.short()` methods are (correctly) referenced to z0_port, which has been set to
    # ref_z0.
    medium, line = keysight_calkit_offset_line(freq, offset_delay, offset_loss, offset_z0, ref_z0)
    if l0 or l1 or l2 or l3:
        poly = np.poly1d([l3, l2, l1, l0])
        inductance = medium.inductor(poly(freq.f)) ** medium.short()
    else:
        inductance = medium.short()
    return line ** inductance


def keysight_calkit_load(freq, offset_delay=0, offset_loss=0, offset_z0=50, ref_z0=50):
    medium, line = keysight_calkit_offset_line(freq, offset_delay, offset_loss, offset_z0, ref_z0)
    ideal_medium = DefinedGammaZ0(frequency=freq, z0=ref_z0)
    load = ideal_medium.match()
    return line ** load


def keysight_calkit_thru(freq, offset_delay=0, offset_loss=0, offset_z0=50, ref_z0=50):
    medium, line = keysight_calkit_offset_line(freq, offset_delay, offset_loss, offset_z0, ref_z0)
    thru = medium.thru()
    return line ** thru
````

## Rohde & Schwarz / 安立信 系数格式在 Rohde & Schwarz 和安立信 VNA 上，使用一种略有不同的格式来定义系数。以下是一个示例，展示了 Maury Microwave 8050CK10，这是一款 3.5 毫米、DC 到 26.5 GHz 校准套件，其格式由 Rohde & Schwarz 定义。[[19](SOLT%20Calibration%20Standards%20Creation.html#ref19)]| 参数 | 单位 | 开路 | 短路 | 负载 | 直通 ||---|---|---|---|---|---|| $\text{C}_0$ | $10^{-15} \text{ F}$ | 62.54 |  |  |  || $\text{C}_1$ | $10^{-15} \text{ F/GHz}$ | 1284.0 |  |  |  || $\text{C}_2$ | $10^{-15} \text{ F/GHz}^2$ | 107.6 |  |  |  || $\text{C}_3$ | $10^{-15} \text{ F/GHz}^3$ | -1.886 |  |  |  || $\text{L}_0$ | $10^{-12} \text{ H}$ |  | 0 |  |  || $\text{L}_1$ | $10^{-12} \text{ H/GHz}$ |  | 0 |  |  || $\text{L}_2$ | $10^{-12} \text{ H/GHz}^2$ |  | 0 |  |  || $\text{L}_3$ | $10^{-12} \text{ H/GHz}^3$ |  | 0 |  |  || 电阻 | $\Omega$ |  |  | 50 |  || 偏移长度 | mm | 4.344 | 5.0017 | 0 | 17.375 || 偏移损耗 | $\text{dB / }\sqrt{\text{GHz}}$ | 0.0033 | 0.0038 | 0 | 0.0065 |### 建模偏移传输线如上所示，本质上是相同的电路模型，唯一的区别在于，偏移传输线以不同的测量单位定义：偏移延迟被定义为物理长度，而不是时间延迟；偏移损耗以分贝为单位定义。偏移 $Z_0$ 被定义为系统阻抗 $Z_\text{ref}$（50 Ω 或 75 Ω），因此未列出。.. 提示::请注意，在 $S_{11}$ 测量中，信号会双向传输，因此以分贝为单位的 $S_{11}$ 损耗在数值上是所陈述的偏移损耗的两倍。在进行简单的单位转换后，我们可以重用 Keysight 模型中的相同计算公式。[[15](SOLT%20Calibration%20Standards%20Creation.html#ref15)]$$\begin{aligned}Z_\text{off} &= Z_\text{ref} \\t_\text{delay} &= \frac{D \cdot \sqrt{\epsilon_r}}{c_0} \\A_\text{loss} &= \frac{L \cdot Z_0}{t_\text{delay} \cdot 20 \log_{10}{(\mathrm{e})}}\end{aligned}$$其中，$D$ 和 $L$ 是 R&S 模型中的偏移长度（米）和偏移损耗（$\text{dB / }\sqrt{\text{GHz}}$），$t_\text{delay}$ 和 $A_\text{loss}$ 是 Keysight 模型中的偏移延迟（秒）和偏移损耗（$\Omega$ / s），$\epsilon_r$ 是介电常数，根据定义为空气，因此 $\epsilon_r = 1$，而 $c_0$ 是光速。$20 \log_{10}{(\mathrm{e})}$ 这一项是将分贝转换为奈培的转换因子。

````python
def rs_to_keysight(rs_offset_length, rs_offset_loss, offset_z0=50):
    offset_delay = rs_offset_length / skrf.constants.c
    offset_loss = skrf.mathFunctions.db_2_np(rs_offset_loss * offset_z0 / offset_delay)
    return offset_delay, offset_loss
````

完成单位转换后，我们可以像定义 Keysight 风格系数中的校准标准一样定义标准。

## 附录：偏移传输线的 $\gamma$ 和 $Z_c$ 的推导### RLCG 线参数的推导根据 Keysight [[4](SOLT%20Calibration%20Standards%20Creation.html#ref4)]，公式 2B.3，*偏移*传输线的参数是相对于物理同轴线定义的，使用以下关系：$$\begin{aligned}A_\text{loss} &= \dfrac{R_0 v}{\sqrt{\epsilon_r \frac{f}{\text{1 GHz}}}} \\t_\text{delay} &= \dfrac{l \sqrt{\epsilon_r}}{v} = \sqrt{L_0 C} \\Z_\text{off} &= \frac{\mu_0 v}{2 \pi \sqrt{\epsilon_r}} \ln(\dfrac{D}{d}) = \sqrt{\frac{L_0}{C}}\end{aligned}$$其中 $v = c_0$ [[13](SOLT%20Calibration%20Standards%20Creation.html#ref13)] 是光速，$\mu_0$ 是真空磁导率，$\epsilon_r$ 是介质的相对介电常数，$R_0$ 和 $L_0$ 是分布的名义电阻和理想导体电感，$C$ 是分布电容，$D$ 和 $d$ 是同轴线的外部和内部直径。通过代数运算：$$\begin{aligned}\left( R_0 \sqrt{\dfrac{f}{10^9}} \right) l &= A_\text{loss} t_\text{delay} = \dfrac{R_0 v}{\sqrt{\epsilon_r \frac{f}{10^9}}} \dfrac{l \sqrt{\epsilon_r}}{v} \\L_0 &= t_\text{delay}  Z_\text{off} = \sqrt{L_0 C} \sqrt{\frac{L_0}{C}}  \\C &= \dfrac{t_\text{delay}}{Z_\text{off}} = \dfrac{\sqrt{L_0 C}}{\sqrt{\frac{L_0}{C}}} \\\end{aligned}$$最后，应用 [[4](SOLT%20Calibration%20Standards%20Creation.html#ref4)]，公式 2B.2 中的定义：$$\begin{aligned}G &= 0 \\L &= L_0 + \dfrac{R}{\omega}\end{aligned}$$完整的 RLCG 线参数的推导如下：$$\begin{aligned}R(f) l &= \left(A_\text{loss} t_\text{delay}\right) \cdot \sqrt{\frac{f}{10^9}} \\L(f) &= L_0 + L_\text{cond} = t_\text{delay}  Z_\text{off} + \dfrac{R}{2\pi f} \\C &= \dfrac{t_\text{delay}}{Z_\text{off}} \\G &= 0 \\l &= 1\text{ (归一化)}\end{aligned}$$### Keysight 的低损耗近似可以使用 RLCG 参数中的通用公式来计算线常数。$$\begin{aligned}Z_c &= \sqrt{ \frac{R + j \omega L}{G + j \omega C}} \\\gamma &= \sqrt{(R + j \omega L) (G + j \omega C)}\end{aligned}$$这是 scikit-rf 的 `DistributedCircuit` 类中常用的方法。然而，Keysight 的官方出版物 [[4](SOLT%20Calibration%20Standards%20Creation.html#ref4)] 和其他文献 [[15](SOLT%20Calibration%20Standards%20Creation.html#ref15)][[20](SOLT%20Calibration%20Standards%20Creation.html#ref20)] 使用了低损耗近似来计算线常数。我们同时展示了公式和代码，以手动计算 $\gamma$ 和 $Z_c$，并使用相同的低损耗近似。如果绝对需要与最后一个数字相同的数值，这可能很有用。将 $\gamma$ 表达式中的项重新组合（参见 [[21](SOLT%20Calibration%20Standards%20Creation.html#ref21)]，公式 2.83）：$$\begin{aligned}\gamma &= \sqrt{(R + j \omega L) (G + j \omega C)} \\       &= j\omega \sqrt{LC} \sqrt{1 - j \left(\frac{R}{\omega L} + \dfrac{G}{\omega C} \right) - \frac{RG}{\omega^2 LC}}\end{aligned}$$当 $G=0$ 时：$$\begin{aligned}Z_c &= \sqrt{ \frac{R + j \omega L}{j \omega C}} \\\gamma &= j\omega \sqrt{LC} \sqrt{1 - j \left(\frac{R}{\omega L}\right)}\end{aligned}$$由于 $L = L_0 + \frac{R}{\omega}$：$$\begin{aligned}Z_c &= \sqrt{ \frac{R + j \omega \left( L_0 + \frac{R}{\omega} \right) }{j \omega C}} \\    &= \sqrt{\dfrac{L_0}{C}} \sqrt{1 + \dfrac{R}{\omega L_0} (1 - j)} \\\gamma &= j\omega \sqrt{\left( L_0 + \frac{R}{\omega} \right) C} \sqrt{1 - j \left(\frac{R}{\omega \left( L_0 + \frac{R}{\omega} \right)}\right)} \\       &= j\omega \sqrt{L_0 C} \sqrt{1 + \dfrac{R}{\omega L_0} (1 - j)}\end{aligned}$$应用一阶泰勒展开 $\sqrt{1 + x} = 1 + x/2$ [[20](SOLT%20Calibration%20Standards%20Creation.html#ref20)]：$$\begin{aligned}\gamma &= j \omega \sqrt{L_0 C} \left[1 + (1 - j) \dfrac{R}{2\omega L_0}\right] \\Z_c &= \sqrt{\dfrac{L_0}{C}} \left[1 + (1 - j) \dfrac{R}{2 \omega L_0}\right]\end{aligned}$$通过将 $R$、$L_0$、$C$ 代入这些方程，最终可以得到 [[4](SOLT%20Calibration%20Standards%20Creation.html#ref4)] 中的官方公式：$$\begin{aligned}\alpha l &= \dfrac{A_\text{loss} t_\text{delay}}{2 Z_\text{off}} \sqrt{\dfrac{f}{10^9}} \\\beta l &= \omega t_\text{delay} + \alpha l \\\gamma l &= \alpha l + j \beta l \\Z_c &= Z_\text{off} + (1 - 1j) \left(\frac{A_\text{loss}}{4 \pi f}\right) \sqrt{\frac{f}{10^9}}\end{aligned}$$### 实现以下代码实现了 Keysight 用于计算线常数的方法。```pythondef offset_gamma_and_zc(offset_delay, offset_loss, offset_z0):    alpha_l = (offset_loss * offset_delay) / (2 * offset_z0)    alpha_l *= np.sqrt(freq.f / 1e9)    beta_l = 2 * np.pi * freq.f * offset_delay + alpha_l    gamma_l = alpha_l + 1j * beta_l    zc = (offset_z0) + (1 - 1j) * (offset_loss / (4 * np.pi * freq.f)) * np.sqrt(freq.f / 1e9)    return gamma_l, zc```### 代码片段将以下定义替换到上面的代码片段中的 `keysight_calkit_offset_line()` 中，以重现完全相同的 Keysight 数据：```pythondef keysight_calkit_offset_line(freq,
