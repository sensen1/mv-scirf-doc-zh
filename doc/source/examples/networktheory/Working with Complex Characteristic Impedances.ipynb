# 处理复数特性阻抗本节讨论在具有可能复数特性阻抗的传输线上电磁波的传播。__要点：__ `scikit-rf` 默认使用“功率波”定义。除非存在具有复数值的特性阻抗，否则这种散射参数定义不应影响您的结果。但是，请注意，“功率波”定义在处理具有复数特性阻抗（例如，对于有损传输线）时，可能会导致不正确的结果/解释。## 特性阻抗与传输线（或任何支持电磁波传播的连续介质）相关的_特性阻抗_ $Z_0$ 被定义为（正向）电压和电流之比，当传输线为无限长时（即 SWR=1，表示没有来自负载的反射，因此没有反向电压和电流）。因此，它描述了线路对电压变化的阻抗，以及对电流变化的阻抗，反之亦然。请注意，在大多数情况下，它_不是_可以使用直流欧姆表[1]测量的阻抗。<img src="working_with_complex_char_impedance_def_char_imp.svg">电压和电流是从基本的电场 $E$ 和磁场 $H$ 派生的量（通过对这些场在传输线横截面上进行线积分，以获得给定的边界条件）。因此，电压和电流取决于传输线的几何形状和材料，并且通常它们的比率不等于 $E/H$ 的比率。对于支持 TEM 以外的其他模式的传输线，电压和电流的定义甚至可能不是唯一的（例如，对于矩形波导）。为了完整性，有时将电场与磁场的比率称为_本征阻抗_（例如，在自由空间中）。因此，$Z_0$ 依赖于材料的特性，并且可以从传输线的分布式电路参数（R、L、C、G）确定。在无损传输线中，$Z_0$ 是实数且为正数。在大多数情况下，即使对于低损耗线路（如同轴电缆[7]）使用实数参考阻抗也是可以的。但是，在其他情况下，例如有损线路、微带线和共面波导，$Z_0$ 可以是复数。旨在实现最大功率传输的匹配问题也可以使用参考复数阻抗的散射参数[5]。## 散射参数定义当需要处理具有复数参考阻抗时，存在三种类型的波：_行波、伪波和功率波_[2-3]。所有这些定义对于具有实数值的特性阻抗都给出相同的结果。

### 行波行波是“真实”的波，因为它们与麦克斯韦方程直接相关，并且可以进行测量。例如，“通过观察槽线实验中入射波和反射波的干涉所产生的驻波的电场峰值和谷值，可以测量行波反射系数”[2-3]。VNA校准中的通过-反射-线 (TRL) 法也测量行波[2]。对于行波，必须知道特性阻抗 $Z_0$，才能确定微波电路理论中的等效电路电压和电流。但是，在实践中，并非总是可以确定 $Z_0$，特别是对于印刷传输线。这就是开发伪波的原因。

### 伪波伪波在 [2-3] 中被引入，作为解决有损传输线中行波复杂性问题的一种方法。虽然伪波是一种数学工具，但它遵循与无损传输线相同的规则和公式。它们的定义如下：$$a = \frac{  \sqrt{\Re(Z_{ref})} }{ 2|Z_{ref}| } \left(V + Z_{ref} I \right)$$$$b = \frac{  \sqrt{\Re(Z_{ref})} }{ 2|Z_{ref}| } \left(V - Z_{ref} I \right)$$其中，$Z_{ref}$ 是伪波的参考阻抗。当 $Z_{ref}=Z_0$ 时，伪波等于行波，其中 $Z_0$ 是传输线的特性阻抗。但是，如果 $Z_{ref}$ 是复数，则伪波通过传输线传输的功率 $p$ 为：$$p = |a|^2 - |b|^2 + 2 \Im(ab^*) \frac{\Im(Z_{ref})}{\Re(Z_{ref})}$$这与行波的结果相同 [2]。此外，如果 $Z_{ref}$ 是复数，则互易器件的散射矩阵可能不是对称的。但是，当选择 $Z_{ref}$ 为实数时，$a$ 和 $b$ 的行为方式与具有实数特性阻抗的无损传输线中的行波幅度相同 [2-3,5]。

### 功率波如上所述，散射矩阵存在一个基本问题，即零反射系数并不能保证从信号源到负载的最大功率传输[5-6]。为了解决这个问题，引入了_功率波_的概念[4]。它是一种与共轭匹配条件兼容的数学工具。它还提供了直观的总功率表达式，即正向波和反向波之和，适用于复杂的特性阻抗。功率波 $a$ 和 $b$ 定义为频率相关的（复数）端口阻抗 $Z_0$ [4]：$$a = \frac{1}{2\sqrt{\Re(Z_0)}} \left( V + Z_0 I \right)$$$$b = \frac{1}{2\sqrt{\Re(Z_0)}} \left( V - Z_0^* I \right)$$其中，$Z_0^*$ 是 $Z_0$ 的复共轭。与伪波不同，功率波满足功率关系：$$p = |a|^2 - |b|^2$$即使在 $Z$ 是复数的情况下。也就是说，功率波被定义为，例如，零功率波反射系数对应于最大功率传输。大多数射频电路求解器都使用功率波的定义（例如 ADS、ANSYS Circuit）。`scikit-rf` 默认也使用功率波的定义。

## 注意事项

### 反射系数与史密斯图

史密斯圆图基于反射系数 $\Gamma$，它是负载阻抗 $Z_{load}$ 和参考阻抗 $Z_{ref}$ 的比值，用于定义圆图的中心：$$\Gamma(Z_{load}) = \frac{Z_{load} - Z_{ref} }{Z_{load} + Z_{ref}}$$

让我们考虑一个短路，对于伪波和功率波两种模型，相对于一个复数特性阻抗，它的反射系数是多少：

```python
%matplotlib inline
import skrf as rf

rf.stylely()
# this is just for plotting junk
kw = dict(draw_labels=True, marker = 'o', markersize = 10)
```

对于伪波的情况，比值 $b/a$ 给出了 $\Gamma$ 的预期表达式。因此，无论参考阻抗如何，短路的史密斯图都如人们所预期的那样：

```python
# same but paying attention to the s-param definition:
ntw_pseudo = rf.instances.wr10.short()
ntw_pseudo.renormalize(1j, s_def='pseudo')
ntw_pseudo.plot_s_smith(**kw)
```

但是，对于功率波：

```python
# default: power wave formulation
ntw_power = rf.instances.wr10.short()
ntw_power.renormalize(1j)  # same as renormalize(1j, s_def='power')
ntw_power.plot_s_smith(**kw)  # gives an open!
```

这种意想不到的行为是由于反向功率波幅值 $b$ 的定义中出现了 $Z_0^*$，从而改变了反射系数公式：$$\Gamma_{\mathrm{powerwaves}}(Z_{load}) = \frac{b}{a} = \frac{Z_{load} - Z_0^*} {Z_{load} + Z_0}$$因此，对于复数特性阻抗的情况，史密斯图不适用于功率波。
