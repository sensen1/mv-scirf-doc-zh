# 安装

## anaconda（推荐）安装 **scikit-rf** 最简单的方法是先安装 [anaconda](http://continuum.io/downloads)，它是一个用于 Python 模块的二进制包管理器。安装完 anaconda 后，您可以通过在终端中输入以下命令来安装 scikit-rf（请勿忘记 `-c conda-forge` 选项）：```conda install -c conda-forge scikit-rf```

## pip使用 [Python 包索引 (PyPi)](https://pypi.org/)，您可以使用以下命令安装 **scikit-rf**：```pip install scikit-rf```

## git您可以从我们的 [github 页面](https://github.com/scikit-rf/scikit-rf) 安装最新版本的 **scikit-rf**。您可以使用 git 直接安装，也可以下载一个 zip 文件。```git clone git@github.com:scikit-rf/scikit-rf.gitcd scikit-rfpython setup.py install```

## 可选模块在安装额外的模块之前，scikit-rf 的某些功能将不可用。您可以使用 conda 或 pip 来安装这些模块。

### 仪器控制要使用 vi 模块，必须安装所有依赖项。可以通过以下命令完成：$ pip install scikit-rf[visa]这将安装以下内容：- [pyvisa](https://pyvisa.readthedocs.io/en/latest/)- [pyvisa-py](https://pyvisa.readthedocs.io/projects/pyvisa-py/en/latest/)
