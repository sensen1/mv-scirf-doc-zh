# 安装

## anaconda（推荐）安装 **scikit-rf** 最简单的方法是先安装 [anaconda](http://continuum.io/downloads)，它是一个用于 Python 模块的二进制软件包管理器。安装完 anaconda 后，您可以通过在终端中输入以下命令来安装 scikit-rf（请不要忘记使用 `-c conda-forge` 选项）：    conda install -c conda-forge scikit-rf

## pip您可以使用 [Python 包索引 (PyPi)](https://pypi.org/) 通过以下命令安装 **scikit-rf**：```pip install scikit-rf```

## git您可以从我们的 [github 页面](https://github.com/scikit-rf/scikit-rf) 安装 **scikit-rf** 的最新开发版本。您可以使用 git 直接安装，也可以下载一个 zip 文件。```git clone git@github.com:scikit-rf/scikit-rf.gitcd scikit-rfpython setup.py install```

## 可选模块在安装额外的模块之前，scikit-rf 的某些功能将不可用。您可以使用 conda 或 pip 来安装这些模块。

### 仪器控制要使用 vi 模块，您必须安装所有依赖项。这可以通过以下命令完成：$ pip install scikit-rf[visa]这将安装以下内容：- [pyvisa](https://pyvisa.readthedocs.io/en/latest/)- [pyvisa-py](https://pyvisa.readthedocs.io/projects/pyvisa-py/en/latest/)

<div class="alert alert-warning">    关于仪器通信的警告    上述内容应该足以与许多仪器进行通信。但是，我们中的一些人遇到了一些难以解决的问题，这些问题与某些仪器配置有关，这些配置需要特定的设置。例如，要通过 GPIB 与 Keysight 的 N5227B VNA 进行通信，我们必须安装    <a href="https://www.keysight.com/us/en/lib/software-detail/computer-software/io-libraries-suite-downloads-2175637.html">Keysight IO Libraries Suite</a>。    在使用虚拟仪器模块时，您可以选择指定 PyVISA 用于与仪器通信的后端。默认情况下，这是完全由 Python 实现的 VISA，即 PyVISA-py，但在上述 Keysight 分析仪的案例中，我们必须指定 VISA 驱动程序的精确 .dll 文件。有关 VISA 后端的更多信息，请参见    [此处](https://pyvisa.readthedocs.io/en/latest/advanced/backends.html)，但如果您遇到问题，请提交详细的错误报告！</div>
