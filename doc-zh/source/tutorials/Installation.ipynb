# 安装

## anaconda（推荐）

安装 **scikit-rf** 最简单的方法是首先安装 
[anaconda](http://continuum.io/downloads)，这是一个用于 python 模块的二进制包管理器。安装好 anaconda 后，你可以在终端中输入以下命令来安装 scikit-rf（不要忘记 `-c conda-forge` 选项）：

    conda install -c conda-forge  scikit-rf

## pip

使用 [Python 包索引 (PyPi)](https://pypi.org/)，你可以通过以下方式安装 **scikit-rf**：

    pip install scikit-rf

## git

**scikit-rf** 的最新开发版本可以从我们的 [github 页面](https://github.com/scikit-rf/scikit-rf) 安装。你可以直接使用 git，或者下载 zip 文件。

    git clone git@github.com:scikit-rf/scikit-rf.git
    cd scikit-rf
    python setup.py install

## 可选模块

scikit-rf 的某些功能需要安装额外的模块才能使用。
你可以使用 conda 或 pip 来安装这些模块。

### 仪器控制

要使用 vi 模块，你必须安装其所有依赖项。这可以通过以下命令完成：

$ pip install scikit-rf[visa]

这将安装以下内容：

- [pyvisa](https://pyvisa.readthedocs.io/en/latest/)
- [pyvisa-py](https://pyvisa.readthedocs.io/projects/pyvisa-py/en/latest/)

<div class="alert alert-warning">
    关于仪器通信的警告

    上述配置应该足以与许多仪器进行通信。然而，
    我们中的一些人在某些仪器配置上遇到了难以解决的问题，
    这些配置需要特定的设置。例如，要通过 GPIB 与 Keysight 的 N5227B PNA 通信，
    我们必须安装 
    <a href="https://www.keysight.com/us/en/lib/software-detail/computer-software/io-libraries-suite-downloads-2175637.html">Keysight IO 库套件</a> 

    在使用虚拟仪器模块时，你可以选择指定 PyVISA 用于与仪器通信的后端。
    默认情况下，这是 PyVISA-py，即完整的 Python VISA 实现，但在上述 Keysight 分析仪的情况下，
    我们必须指定 VISA 驱动程序的确切 .dll 文件。有关 VISA 后端的更多信息可以在
    [这里](https://pyvisa.readthedocs.io/en/latest/advanced/backends.html) 找到，
    但如果你遇到问题，请提交详细的错误报告！
</div>
