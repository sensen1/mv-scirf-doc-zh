# 校准

本教程介绍 `skrf` 中的校准功能。校准是将原始测量数据转换为校正数据的过程。

## 校准类型

`skrf` 支持多种校准算法：

- **SOLT**: 短路-开路-负载-传输
- **TRL**: 传输-反射-线
- **Multiline TRL**: 多线 TRL
- **NIST Multiline TRL**: NIST 多线 TRL
