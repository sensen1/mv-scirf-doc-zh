# 绘图

`skrf` 提供了多种绘图功能来可视化网络数据。本教程介绍这些功能。

## 导入

```python
import skrf as rf
import matplotlib.pyplot as plt

rf.style()
```

## 基本绘图

网络对象有各种绘图方法。最常见的包括：

```python
ntwk = rf.data.ring_slot
ntwk.plot_s_mag()
```

### 幅度图

```python
ntwk.plot_s_mag(m=0, n=0)  # S11
ntwk.plot_s_mag(m=1, n=0)  # S21
```

### 相位图

```python
ntwk.plot_s_deg()
```

### 史密斯圆图

```python
ntwk.plot_s_smith()
```

## 复平面绘图

```python
ntwk.plot_s_complex()
```

## 时域绘图

```python
ntwk.plot_s_db_time()
```

## 自定义绘图

您可以使用 matplotlib 的功能来自定义绘图：

```python
fig, ax = plt.subplots()
ntwk.plot_s_mag(ax=ax)
ax.set_title('自定义标题')
ax.set_xlabel('频率 (GHz)')
```

## 多网络绘图

```python
ntwk1 = rf.data.ring_slot
ntwk2 = rf.data.line

fig, ax = plt.subplots()
ntwk1.plot_s_mag(ax=ax, label='环形缝隙')
ntwk2.plot_s_mag(ax=ax, label='传输线')
ax.legend()
```

## 结论

`skrf` 的绘图功能使网络数据的可视化变得简单。更多高级绘图选项请参阅 API 文档。
