# 网络

本教程介绍 `skrf` 中的 [Network](../api/network.rst) 对象及其基本功能。

## 创建网络

可以通过多种方式创建网络：

### 从 Touchstone 文件

最常见的方式是从 Touchstone 文件（.sNp）创建网络。

```python
import skrf as rf
ntwk = rf.Network('data/ring slot.s2p')
```

### 从 s 参数直接创建

也可以直接从 s 参数数组创建网络。

```python
import numpy as np
freq = rf.Frequency(1, 10, 101, 'ghz')
s = np.random.rand(101, 2, 2) + 1j*np.random.rand(101, 2, 2)
ntwk = rf.Network(frequency=freq, s=s, name='random_2port')
```

## 网络属性

网络对象有许多有用的属性：

```python
# 散射参数矩阵
ntwk.s.shape
```

```python
# 端口阻抗
ntwk.z0
```

```python
# 频率对象
ntwk.frequency
```

## 网络操作

网络支持各种数学运算：

```python
# 级联
ntwk1 = rf.data.ring_slot
ntwk2 = rf.data.line
cascade = ntwk1 ** ntwk2
```

```python
# 去嵌入
deembedded = ntwk1.inv ** ntwk2
```

## 保存网络

网络可以保存为 Touchstone 格式：

```python
ntwk.write_touchstone('output.s2p')
```
