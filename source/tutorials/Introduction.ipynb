## 网络

`skrf` 中的核心对象是 N 端口微波 [Network][Network] 对象。可以通过多种方式创建 [Network][Network]，其中一种方式是从存储在 touchstone 文件中的数据创建。

[Network]: ../api/network.rst

```python
ring_slot = rf.Network('data/ring slot.s2p')
```

如果找不到 `ring slot.s2p`，则只需从 `skrf.data` 模块导入它。


```python
from skrf.data import ring_slot  # noqa: F811
```

如果在命令行中输入，将打印出网络的简短描述

```python
ring_slot
```

微波 [Network](../api/network.rst) 的基本属性由以下属性提供，


* `Network.s` : 散射参数矩阵。
* `Network.z0`  : 端口阻抗矩阵。
* `Network.frequency`  : 频率对象。

 [Network](../api/network.rst) 对象还有许多其他属性和方法，可以在其文档字符串中找到。如果您使用 IPython/Jupyter，则可以在命令行上通过 'tab' 键调出这些属性和方法。

	ring_slot.s              ring_slot.s_arcl         ring_slot.s_im
	ring_slot.s11            ring_slot.s_arcl_unwrap  ring_slot.s_mag

构建网络的其他方法详见 [网络教程](Networks.ipynb)。

### 线性操作

	
可以通过重载的运算符访问对 s 参数的逐元素数学操作。为了说明，我们从 `skrf.data` 模块加载几个 `Networks`。

```python
short = rf.data.wr2p2_short
delayshort = rf.data.wr2p2_delayshort
```

它们 s 参数之间的复数差计算为

```python
short - delayshort
```

这返回一个新的 [Network](../api/network.rst)。其他算术运算符也被重载，

```python
short/delayshort
```

### 级联和去嵌入

级联网络由 `**` 运算符执行。

```python
short = rf.data.wr2p2_short
line = rf.data.wr2p2_line
```

```python
short ** line ** short
```

去嵌入本质上与级联相同，但使用逆网络。`Network.inv` 属性返回一个网络的逆，

```python
line.inv ** short ** line.inv
```

有关级联和去嵌入的更详细讨论，请参阅 [连接网络教程](Connecting_Networks.ipynb)。
