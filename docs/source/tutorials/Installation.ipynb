# 安装

scikit-rf 支持 Python 3.8+。

## 使用 pip

最简单的方法是使用 pip 安装，

```python
pip install scikit-rf
```

## 使用 conda

或者，如果您使用 conda，

```python
conda install -c conda-forge scikit-rf
```

## 可选依赖

要获得完整功能，您可能还想安装以下可选依赖：

```python
pip install scikit-rf[plot,xlsx,netw,visa]
```

这将安装绘图、Excel I/O、网络分析和仪器控制的支持。

## 从源码安装

如果您想从源码安装，可以从 GitHub 克隆仓库，

```python
git clone https://github.com/scikit-rf/scikit-rf.git
cd scikit-rf
pip install -e .
```

## 验证安装

安装后，您可以通过导入包来验证安装，

```python
import skrf as rf
print(rf.__version__)
```
